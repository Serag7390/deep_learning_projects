```python
import nltk

nltk.download("punkt")

nltk.download("wordnet")

nltk.download("stopwords")

nltk.download('omw-1.4')

from nltk.corpus import stopwords

from nltk.stem import WordNetLemmatizer 

import string
```

```python
vocab = []

classes = []

doc_X = []

doc_y = []

for intent in data["intents"]:
    
    for pattern in intent["patterns"]:
        
        tokens = nltk.word_tokenize(pattern)
        
        vocab.extend(tokens)
        
        doc_X.append(pattern)
        
        doc_y.append(intent["class"])
    
    if intent["class"] not in classes:
        classes.append(intent["class"])
```

```python
vocab = [word for word in vocab if word not in string.punctuation]

lemmatizer = WordNetLemmatizer()

vocab = [lemmatizer.lemmatize(word.lower() ) for word in vocab]

stop_words = stopwords.words('english')
vocab = [word for word in vocab if word not in stop_words]

vocab = set(vocab)

vocab = sorted(vocab)
vocab
```

```python
classes = sorted(set(classes))
classes
```

```python
import random 
import numpy as np

X = []

y= []

for idx, doc in enumerate(doc_X):
    bow = []
    
    doc = lemmatizer.lemmatize(doc.lower())
    
    for word in vocab:
        bow.append(1) if word in doc else bow.append(0)
    
    X.append(bow)
    
    output_row = [0] * len(classes)

    cl = doc_y[idx]
    
    cl_index= classes.index(cl)
    output_row[cl_index] = 1
    
    y.append(output_row)
```

```python
from keras import Sequential 
from keras.layers import Input, Dense, Dropout

input_shape = (len(X[0]),)

output_shape = len(y[0])

model = Sequential()
model.add(Input(shape=input_shape))
model.add(Dense(128, activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(64, activation="relu"))
model.add(Dropout(0.3))
model.add(Dense(output_shape, activation = "softmax"))
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=["accuracy"])
```

```python
import numpy as np

X_training_np = np.array(X)
y_training_np = np.array(y)
```

```python
model.fit(X_training_np, y_training_np, epochs=200)
```

```python
model.save('chatbot_model.keras')

import pickle

vocab_path = 'chatbot_vocab'
with open(vocab_path, 'wb') as file:
    pickle.dump(vocab, file)

classes_path = 'chatbot_classes'
with open(classes_path, 'wb') as file:
    pickle.dump(classes, file)
```

```python
import nltk

def clean_text(text):
    
    tokens = nltk.word_tokenize(text)
    tokens = [lemmatizer.lemmatize(word) for word in tokens ]
    
    stop_words = stopwords.words('english')
    
    tokens = [word for word in tokens if word not in stop_words]
    
    return tokens


def bag_of_words(text, vocab): 
    
    tokens = clean_text(text)
    bow = [0] * len(vocab)
    for w in tokens: 
        for idx, word in enumerate(vocab):
            if word == w: 
                bow[idx] = 1
    return bow

def pred_class(text, model, vocab, classes): 
    
    bow = bag_of_words(text, vocab)
    predictions = model.predict([bow], verbose=0)[0]
    
    max_index=np.argmax(predictions)
    
    return classes[max_index]

def get_response(pred_cl, intents_json): 
    list_of_intents = intents_json["intents"]
    for x in list_of_intents: 
        if x["class"] == pred_cl:
            result = random.choice(x["responses"])
            break
    return result
```

```python
import json

from keras.models import load_model
import pickle

def chatbot(intents_path, model_path, vocab_path, classes_path): 
    with open(intents_path) as file:
        data = json.load(file)

    model = load_model(model_path)

    with open(vocab_path, 'rb') as file:
        vocab = pickle.load(file)

    with open(classes_path, 'rb') as file:
        classes = pickle.load(file)

    message = input("")
    while len(message)>0:
        message= message.lower()
        pred_cl = pred_class(message, model, vocab, classes)
        result = get_response(pred_cl, data)
        print(result)
        message = input("")
```

```python
chatbot('intents.json', 'chatbot_model.keras', 'chatbot_vocab', 'chatbot_classes')
```
